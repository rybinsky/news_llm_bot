```python
from datetime import datetime
from typing import Dict, Any

def extract_article_data(article_dict: dict[str, Any]) -> dict[str, Any]:
    """
    Извлекает и структурирует данные статьи из словаря newspaper.
    
    Args:
        article_dict: Словарь с данными статьи, полученный из newspaper.Article
    
    Returns:
        Словарь с извлеченными и структурированными данными статьи
    """
    data = {
        'url': article_dict.get('url'),
        'source_url': article_dict.get('source_url'),
        'title': article_dict.get('title'),
        'text': article_dict.get('text'),
        'summary': article_dict.get('summary'),
        'publish_date': article_dict.get('publish_date'),
        'language': article_dict.get('meta_lang'),
        'keywords': article_dict.get('keywords', []) + article_dict.get('meta_keywords', []),
    }
    
    if data['publish_date'] and isinstance(data['publish_date'], datetime):
        data['publish_date'] = data['publish_date'].isoformat()
    
    return data
```

```python
from langchain_ollama.llms import OllamaLLM
from langchain.prompts import FewShotPromptTemplate, PromptTemplate

# Заранее заданные категории
categories = ["Политика", "Наука", "Спорт", "Технологии", "Культура"]

# Примеры текстов и их тематик (few-shot learning)
examples = [
    {"text": "Ученые открыли новый вид динозавров", "category": "Наука"},
    {"text": "Выборы президента пройдут в следующем месяце", "category": "Политика"},
    {"text": "Футбольный клуб нашего города выиграл чемпионат!", "category": "Спорт"},
]

# Шаблон для каждого примера
example_template = """
Текст: {text}
Тема: {category}
"""

# Создаем PromptTemplate для примеров
example_prompt = PromptTemplate(
    input_variables=["text", "category"],
    template=example_template,
)

# Формируем FewShotPromptTemplate
few_shot_prompt = FewShotPromptTemplate(
    examples=examples,
    example_prompt=example_prompt,
    prefix="Определи тематику текста из списка и выведи только ее без других символов: " + ", ".join(categories),
    suffix="Текст: {input}\nТема:",
    input_variables=["input"],
)
model = OllamaLLM(model="llama3.1:8b", temperature=0.1)

input_text = "ИИ превзошел человека в шахматах"
chain = few_shot_prompt | model

chain.invoke({'input': "Почему долла стал падать???"})
```

```python
from sqlalchemy import Column, DateTime, Integer, String, Text
from sqlalchemy.dialects.postgresql import JSONB, ARRAY as PG_ARRAY
from sqlalchemy.ext.declarative import declarative_base


Base = declarative_base()


class NewsArticle(Base):
    __tablename__ = "news_articles"

    id = Column(Integer, primary_key=True)
    title = Column(String(500))
    text = Column(Text)
    topic = Column(String(50), index=True)
    publish_date = Column(DateTime)
    url = Column(String(1000), unique=True)
    source = Column(String(200), index=True)
    keywords = Column(PG_ARRAY(String(50)))
    summary = Column(Text)
    language = Column(String(10))
    embedding = Column(JSONB)


    def __repr__(self):
        return f"<NewsArticle(title='{self.title}', source='{self.source}')>"
```

```python
def init_db():
    """Initialize the database by creating tables"""
    Base.metadata.create_all(bind=engine)

init_db()
```
